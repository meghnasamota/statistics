# Measure Of Tendency

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv(r'titanic.csv')
df.head(3)
```

# mean

```python
mn=df['Age'].mean()
mn
```

```python
mn1=df['Fare'].mean()
mn1
```

```python
sns.histplot(x='Age',data=df,bins=[i for i in range(0,81,10)])
plt.plot([mn for i in range(0,300)],[i for i in range(0,300)],c='red')
```

```python
sns.histplot(x='Fare',data=df,bins=[i for i in range(0,81,10)])
plt.plot([mn1 for i in range(0,300)],[i for i in range(0,300)],c='red')
```

# median

```python
 df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
md=df['Age'].median()
md
```

```python
sns.histplot(x='Age',data=df,bins=8)
plt.plot([md for i in range(0,300)],[i for i in range(0,300)],color='red')
```

```python
md1=df['Fare'].median()
md1
```

```python
sns.histplot(x='Age',data=df,bins=8)
plt.plot([md1 for i in range(0,300)],[i for i in range(0,300)],color='red')
```

# mode

```python
mo=df['Age'].mode()[0]
mo
```

```python
mo1=df['Fare'].mode()[0]
mo1
```

```python
sns.histplot(x='Fare',data=df,bins=[i for i in range(0,81,10)] )
plt.plot([mo1 for i in range(0,300)],[i for i in range(0,300)],color='blue')
plt.show()
```

# mean, mode and median

```python
sns.histplot(x='Fare',data=df,bins=[i for i in range(0,81,10)])
plt.plot([mn1 for i in range(0,300)],[i for i in range(0,300)],color='green',label='mean')
plt.plot([md1 for i in range(0,300)],[i for i in range(0,300)],color='blue',label='median')
plt.plot([mo1 for i in range(0,300)],[i for i in range (0,300)],color='red',label='mode')
plt.legend()
plt.show()
```
